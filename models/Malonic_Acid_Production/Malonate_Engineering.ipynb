```python
# An update to iMT1026 from 2017
# Fine‐tuning the P. pastoris iMT1026 genome‐scale metabolic model for improved prediction of growth on methanol or glycerol as sole carbon sources
model = read_sbml_model('iMT1026-v3.xml')
```

```python
medium = model.medium
with model:
    medium['Ex_meoh'] = 10
    medium['Ex_glyc'] = 0
    model.medium = medium
    model.objective = model.reactions.Ex_biomass
    print(model.optimize().objective_value)
```

```python
from cobra import Reaction, Metabolite
new_reaction = Reaction('Mals') 
Malonate = Metabolite(id='mln_c', compartment='c')
new_reaction.add_metabolites({model.metabolites.msa_c: -1,
                              model.metabolites.nad_c: -1,
                              model.metabolites.h2o_v: -1,
                              Malonate: 1,
                              model.metabolites.nadh_c: 1,
                              model.metabolites.h_c: 2,
                             })
model.add_reactions([new_reaction])
model.reactions.Mals.lower_bound = -1000

```

```python
model.reactions.Mals
```

```python
model.metabolites.msa_c.formula
```

```python
model.metabolites.mln_c.formula = 'C3H2O4'
model.metabolites.mln_c.name = 'Malonate'
```

```python
# Adding the exchange reaction
Malonate_exchange = Reaction('EX_mln_c')
Malonate_exchange.add_metabolites({model.metabolites.mln_c: -1})
model.add_reaction(Malonate_exchange)
model.reactions.EX_mln_c.lower_bound = 0
```

```python
medium = model.medium
medium['Ex_meoh'] = 10
medium['Ex_glyc'] = 0
model.medium = medium
print(model.optimize().objective_value)
```

```python
# Simulating for optimized product flux
model.objective = model.reactions.EX_mln_c
solution = model.optimize()
```

```python
#Yield in mmol_mln*mmol_meoh^-1

solution.fluxes['EX_mln_c']/-solution.fluxes['Ex_meoh']
```

```python
solution.fluxes['Ex_biomass']
```

```python
#Checking at 90% growth rate
with model:
    model.objective = model.reactions.Ex_biomass 
    specified_growth_rate = 0.90*model.optimize().objective_value


    constrained_growth = model.problem.Constraint(
        model.reactions.get_by_id('Ex_biomass').flux_expression,
        lb=specified_growth_rate,
        ub=specified_growth_rate)
    model.add_cons_vars(constrained_growth)

    model.objective = model.reactions.EX_mln_c
    solution = model.optimize()
    print(solution.fluxes['EX_mln_c']/-solution.fluxes['Ex_meoh'])
```

Then we could go on forever knockouting genes and nature would "Find a way" around it. Instead use phenotypic plane to see from the phenotypic phase plane, that the best use of recourses is at growth rate 0.060535

```python
model.reactions.ASPTA.upper_bound = 1000
```

```python
with model:
    constrained_growth = model.problem.Constraint(
        model.reactions.get_by_id('Ex_biomass').flux_expression,
        lb=0.060535,
        ub=0.060535)
    model.add_cons_vars(constrained_growth)

    model.objective = model.reactions.EX_mln_c
    asp_sum = model.metabolites.asp_L_c.summary()
    oaa_sum = model.metabolites.oaa_c.summary()
    msa_sum = model.metabolites.msa_c.summary()
```

It's very interesting how the reaction they are trying to amplify in the papers (OAA to ASP) is causing the exact opposite.

An possible knockout could be this other consuming reaction. Only an enzyme that can convert OAA to PEP is in pichia. Miao expresses ATP forming carboxykinase 4.1.1.49)

```python
with model:
    model.reactions.PPCK.lower_bound=-1000
    model.reactions.PPCK.upper_bound=0

    model.objective = model.reactions.EX_mln_c
    solution = model.optimize()
    print(solution.fluxes['EX_mln_c']/-solution.fluxes['Ex_meoh'])
    asp_sum = model.metabolites.asp_L_c.summary()
    oaa_sum = model.metabolites.oaa_c.summary()
    msa_sum = model.metabolites.msa_c.summary()
    prod_env5 = production_envelope(model, ["Ex_biomass"], objective="EX_mln_c", carbon_sources="Ex_meoh")
```
