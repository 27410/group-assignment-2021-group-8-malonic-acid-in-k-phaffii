# Pathway Evaluation
The goal is to analyse the behaviour of the different pathway when they are fed with the chosen substrate Methanol. They are judged by their maximum production and the maximum MA flux in relation to the biomass growth rate.

```python
# Importing used packages and model
from cobra.io import read_sbml_model
from cobra import Reaction, Metabolite
#import pandas as pd
model = read_sbml_model('iMT1026-v3.xml')
%run -i create_pathway.ipynb
# flux unit is mmol / [gDW h]
```

```python
# path_selector legend
# 0 -> no additional reactions
#1 -> beta alanine pathway  
#2 -> transferase malonyl-CoA pathway  
#3 -> oxidoreuctive malonyl-CoA pathway
#4 -> lactate pathway  
#create_pathway(model,3)  # self-built function create_pathway(model, path_selector), that adds pathway

#factor_growth = 0.5
beta_model = create_pathway(model.copy(),1)
trn_malonyl_model = create_pathway(model.copy(),2)
red_malonyl_model = create_pathway(model.copy(),3)
lactate_model = create_pathway(model.copy(),4)
models = [(beta_model,'Beta-Alanine'),(trn_malonyl_model,'Malonyl-CoA Transferase'),(red_malonyl_model,'Malonyl-CoA Oxidored'),(lactate_model,'Lactate')]
```

```python
# General medium for all models based on the unchanged vanilla model
# High uptake rate to make the differnces in the pathway more visible
medium = model.medium
medium['Ex_meoh'] = 6  # 6 = Realistic methanol uptake rates cf.Tomàs‐Gamisans 2018
medium['Ex_glyc'] = 0
# medium['Ex_btn'] = 10
```

```python
from cobra.flux_analysis.phenotype_phase_plane import production_envelope
# import pandas as pd  # todo try to put all 4 graphs into one plot for better visability
model.medium = medium
model.objective = model.reactions.get_by_id('Ex_biomass')
print('General maximum growth',model.slim_optimize())

for pairing in models:
    pathway_model = pairing[0]
    print(pairing[1])
    with pathway_model:
        #print(medium)
        pathway_model.medium = medium
        pathway_model.objective = pathway_model.reactions.get_by_id('Ex_mln')
        print('Maximum malonic acid production',pathway_model.slim_optimize())
        ppp = production_envelope(pathway_model,
                    reactions='Ex_biomass',
                    objective='Ex_mln')
        ppp.plot(x='Ex_biomass', y='flux_maximum', title=pairing[1])
```

### Discussion
The reslts show, that first of all the maximum yield of malonic acid is the highest for the beta-alanine and malonyl-CoA Transerase pathway. Also the phenotypic phase plane shows a better behaviour in the curvature. For teh first two pathways, a smaller reduction over the increasing biomass growth rate is visible, before having a crack in the curvture and going down to the maximum growth rate with no malonic acid production.  
The results show that the first two pathways are preferable for malonic acid production and ar further examined for strain improvements

# Evaluating optimal O2 uptake

```python
models2 = [models[0],models[1]]
for pairing in models2:
    pathway_model = pairing[0]
    print(pairing[1])
    with pathway_model:
        #print(medium)
        pathway_model.medium = medium
        pathway_model.objective = pathway_model.reactions.get_by_id('Ex_mln')
        print('Maximum malonic acid production',pathway_model.slim_optimize())
        ppp = production_envelope(pathway_model,
                    reactions='Ex_o2',
                    objective='Ex_mln')
        ppp.plot(x='Ex_o2', y='flux_maximum', title=pairing[1])
        print('Optimal O2 uptake', ppp.Ex_o2[ppp.flux_maximum.idxmax()])
```

```python
# and for biomass growth
for pairing in models2:
    pathway_model = pairing[0]
    print(pairing[1])
    with pathway_model:
        #print(medium)
        pathway_model.medium = medium
        pathway_model.objective = pathway_model.reactions.get_by_id('Ex_biomass')
        print('Maximum biomass growth',pathway_model.slim_optimize())
        ppp = production_envelope(pathway_model,
                    reactions='Ex_o2',
                    objective='Ex_biomass')
        ppp.plot(x='Ex_o2', y='flux_maximum', title=pairing[1])
        print('Optimal O2 uptake', ppp.Ex_o2[ppp.flux_maximum.idxmax()])
```

### Discussion
The results for the two pathways for the oxygen dependency are equal. Both the malonic acid production and the biomass growth rate seem to have the same oxygen dependency for both pathways.  
Interestingly the oxygen demand varys between the two objectives malonic acdi producation and biomass growth. The optimal uptake is slightly higher for the biomass growth. Also a difference in the curvature is visible.

# Evaluating for minimal medium

```python
from cobra.medium import minimal_medium
print(medium)
```

```python
# for biomass growth
for pairins in models2:
    pathway_model = pairing[0]
    with pathway_model:
        pathway_model.medium = medium
        max_growth = pathway_model.slim_optimize()
        print(pairing[1], 'with maximum growth of',max_growth)
        print(minimal_medium(pathway_model, max_growth))
```

```python
# for malonic acid production growth
for pairins in models2:
    pathway_model = pairing[0]
    with pathway_model:
        pathway_model.medium = medium
        pathway_model.objective = pathway_model.reactions.get_by_id('Ex_mln')
        max_production = pathway_model.slim_optimize()
        print(pairing[1], 'with maximum growth of',max_production)
        print(minimal_medium(pathway_model, max_production))
```

### Discussion
As seen in the results, the medium only needs a few components to ensure the growth and production. In general the substrate is the limitation in both cases. Further it has to be noticed, that the demands to the medium are much lower for the malonic acid production, which jut needs a c-source and oxygen to work.  
Nevertheless, it has to be noted, that several compounds are deemed as unnecessary which has to be seen critical. First and for most the compley additive Biotin doesnt seem to effect the growth or malic acid production in any way, while it is known to be vital to yeast organisms.
